```python
# Import All the required Libraries
import kaggle
import zipfile
import pandas as pd
import sqlalchemy as sal
```

```python
# download dataset using kaggle API, We can use kaggle api calls to download
!kaggle datasets download ankitbansal06/retail-orders -f orders.csv
```

```python
# We have downloaded the zipfile, we will use python libraries to unzip and extract the required file
zip_ref = zipfile.ZipFile('orders.csv.zip') # We can pass the complete file path here.
zip_ref.extractall() # extract file to dir, we can pass the file location where we want to extract the zipfiles.
zip_ref.close() # close file
```

```python
# Read data from the extracted file and handle the null values

df = pd.read_csv('orders.csv') # can pass complete file path as well

# Looking at the data

df.head(20)

# First figure out what can be treated as null values from our data records. So have a look at the distinct values of ship mode.

df['Ship Mode'].unique()
```

```python
#In our case, we can treat 'Not Available' and 'unknown' as nan values. In pandas 'nan' represents a null values

df = pd.read_csv('orders.csv',na_values=['Not Available', 'unknown'])

# Verify that we handled null values.

df.head(20)

# Check the unique values present in our dataset

df['Ship Mode'].unique()
```

```python
# Rename the columns to standard names.

# First check all the column names we have

df.columns
```

```python
# Renaming can be done in different ways.
# 1. Manually calling the pandas rename() function by passing dictionary in it.

df.rename(columns={'Ship Mode': 'ship_mode', 'Order Id' : 'order_id'})

# 2. Using string functions
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

```python
# Verify the data
df.head(10)
```

```python
# Derive new columns: discount, sales_price, profit

# 1. discount
df['discount']=df['list_price']*df['discount_percent']*.01

# 2. sales_price
df['sale_price']= df['list_price']-df['discount']

# 3. profit
df['profit']=df['sale_price']-df['cost_price']
```

```python
# Verify the data

df.head(10)
```

```python
# Handle the datatypes of each column. convert the datatypes to its corresponding data
df.dtypes
```

```python
# 1. convert order_date datatype to datetime

df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
# Check for correct datatypes now
df.dtypes
```

```python
# Drop unnecessary columns as part of cleaning. Example: 'list_price','cost_price','discount_percent'
df.drop(columns=['list_price','cost_price','discount_percent'],inplace=True)
```

```python
# Verify the data
df.head(20)
```

```python
# Establish a connection with sql server
engine = sal.create_engine('mssql://SAMEER\SQLEXPRESS/master?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')
conn=engine.connect()
```

```python
# load the data into sql server using replace option, replace option will check for the table we passed. if found then it will drop and create the already present table and then replace it with table we passed using to_sql()
# it will create the new table
df.to_sql('df_orders', con=conn , index=False, if_exists = 'replace')

# Issue is that: Pandas will try to provide the maximum memory to all the columns and assigns them maximum values. 
# Example: for string column, it will assign 'varchar(max)'', for numeric columns like quantity, it will assign 'BigInt'. Though it is not required.
# Due to this we use the append option to create tables.
```

```python
# load the data into sql server using append option, append option will check for the table we passed. if found then it will append it with table we passed using to_sql()

# First drop the above created table and create it again using append option

# For this, we can create a table in the sql server priorly.

df.to_sql('df_orders', con=conn , index=False, if_exists = 'append')
```
